# Make smaller datasets out of TableType and Top25Type
The main cleanup that needs to be done is to split the giant dataset into separate smaller datasets. The logic behind this is that each miniature dataset only contains mathematically comparable data. So, for example, one dataset wouldn't contain both medians and frequencies, because those aren't comparable data points. 

```python
# overall and cost combo 
overall_cost = health[(health['TableType'] == 'Overall') & (health['Top25Type'] == 'Cost')]
overall_cost.to_csv('/home/jovyan/final-group27/data/overall_cost.csv', index = False)

# overall and freq combo 
overall_freq = health[(health['TableType'] == 'Overall') & (health['Top25Type'] == 'Freq')]
overall_freq.to_csv('/home/jovyan/final-group27/data/overall_freq.csv', index = False)

# overall and median combo 
overall_med = health[(health['TableType'] == 'Overall') & (health['Top25Type'] == 'Median')]
overall_med.to_csv('/home/jovyan/final-group27/data/overall_med.csv', index = False)

# top 25 list overall and cost
top_cost = health[(health['TableType'] == 'Top 25 List Overall') & (health['Top25Type'] == 'Cost')]
top_cost.to_csv('/home/jovyan/final-group27/data/top_cost.csv', index = False)

# top 25 list overall and freq combo 
top_freq = health[(health['TableType'] == 'Top 25 List Overall') & (health['Top25Type'] == 'Freq')]
top_freq.to_csv('/home/jovyan/final-group27/data/top_freq.csv', index = False)

# top 25 list overall and median combo 
top_med = health[(health['TableType'] == 'Top 25 List Overall') & (health['Top25Type'] == 'Median')]
top_med.to_csv('/home/jovyan/final-group27/data/top_med.csv', index = False)

# top 25 payer and cost
payer_cost = health[(health['TableType'] == 'Top 25 Payer') & (health['Top25Type'] == 'Cost')]
payer_cost.to_csv('/home/jovyan/final-group27/data/payer_cost.csv', index = False)

# top 25 payer and freq combo 
payer_freq = health[(health['TableType'] == 'Top 25 Payer') & (health['Top25Type'] == 'Freq')]
payer_freq.to_csv('/home/jovyan/final-group27/data/payer_freq.csv', index = False)

# top 25 payer and median combo 
payer_med = health[(health['TableType'] == 'Top 25 Payer') & (health['Top25Type'] == 'Median')]
payer_med.to_csv('/home/jovyan/final-group27/data/payer_med.csv', index = False)

# top 25 totals and cost
tot_cost = health[(health['TableType'] == 'Top 25 Totals') & (health['Top25Type'] == 'Cost')]
tot_cost.to_csv('/home/jovyan/final-group27/data/tot_cost.csv', index = False)

# top 25 totals and freq combo 
tot_freq = health[(health['TableType'] == 'Top 25 Totals') & (health['Top25Type'] == 'Freq')]
tot_freq.to_csv('/home/jovyan/final-group27/data/tot_freq.csv', index = False)

# top 25 totals and median combo 
tot_med = health[(health['TableType'] == 'Top 25 Totals') & (health['Top25Type'] == 'Median')]
tot_med.to_csv('/home/jovyan/final-group27/data/tot_med.csv', index = False)

# Top 25 W/O Medi-Cal and cost
medi_cost = health[(health['TableType'] == 'Top 25 W/O Medi-Cal') & (health['Top25Type'] == 'Cost')]
medi_cost.to_csv('/home/jovyan/final-group27/data/medi_cost.csv', index = False)

# Top 25 W/O Medi-Cal and freq combo 
medi_freq = health[(health['TableType'] == 'Top 25 W/O Medi-Cal') & (health['Top25Type'] == 'Freq')]
medi_freq.to_csv('/home/jovyan/final-group27/data/medi_freq.csv', index = False)

# Top 25 W/O Medi-Cal and median combo 
medi_med = health[(health['TableType'] == 'Top 25 W/O Medi-Cal') & (health['Top25Type'] == 'Median')]
medi_med.to_csv('/home/jovyan/final-group27/data/medi_med.csv', index = False)
```
